### Funciones auxiliares

```python
import numpy as np
import pandas as pd
import datetime as dt

from pandas_datareader import data as pdr
import yfinance as yf

import seaborn as sns
import matplotlib.pyplot as plt

def get_coov(act1, act2):
  yf.pdr_override()
  start = dt.datetime(2015,1,1)
  end = dt.datetime.now()

  etflist = [act1, act2]

  data = pdr.get_data_yahoo(etflist, start, end)
  close_price = data['Close']

  # calculando variación procentual
  change_price = close_price.pct_change().dropna()

  # reordenando las columnas para dejar al SPY en la primera
  change_price = change_price[[act1, act2]]
  # matriz de correlación
  corr = change_price.corr()

  return corr.iloc[0][1]


#get_coov('ETH-USD', 'BTC-USD')

```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def graficar_evolucion_total(etiquetas, capital, inicio, fin):
    # Calcular el capital inicial invertido en cada empresa
    capital_por_empresa = capital / len(etiquetas)

    # Crear un DataFrame vacío para almacenar los valores del capital invertido en cada empresa
    valores_capital = pd.DataFrame()

    # Obtener los datos de cotización de Yahoo Finance para cada empresa y calcular el valor del capital invertido
    for etiqueta in etiquetas:
        data = yf.download(etiqueta, start=inicio, end=fin)
        porcentaje_cambio = data['Adj Close'].pct_change()
        valor_capital = capital_por_empresa * (1 + porcentaje_cambio).cumprod()
        valores_capital[etiqueta] = valor_capital

    # Sumar los valores de capital invertido en todas las empresas para obtener el valor total del capital
    valores_capital['Total'] = valores_capital.sum(axis=1)

    # Graficar la evolución del valor total del capital invertido
    valores_capital['Total'].plot()

    # Configurar el título y las etiquetas de los ejes
    plt.title('Evolución del capital invertido en múltiples empresas')
    plt.xlabel('Fecha')
    plt.ylabel('Valor del capital invertido')

    # Mostrar la gráfica
    plt.show()

# # Ejemplo de uso:
# etiquetas = ['AAPL', 'MSFT', 'GOOG']
# capital_inicial = 10000
# inicio = "2015-01-01"
# fin = "2018-12-31"
```

### Ejemplo con 4 assets MUY Correlacionados 2 a 2

#### Dibujar el grafo con las correlaciones entre las empresas

```python
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
import networkx as nx

# Definir los símbolos de las acciones y las fechas de inicio y fin
tickers = ['BBVA', 'SAN', 'GOOG', 'GOOGL']

# Crear el grafo
G = nx.Graph()

# Agregar nodos al grafo con los nombres de las empresas
for i, ticker in enumerate(tickers):
    G.add_node(i, label=ticker)

# Agregar aristas ponderadas al grafo
num_assets = len(tickers)
for i in range(num_assets):
    for j in range(i+1, num_assets):
        weight = get_coov(tickers[i], tickers[j])
        G.add_edge(i, j, weight=weight)

# Dibujar el grafo
pos = nx.spring_layout(G)  # Calcular posiciones de los nodos
nx.draw(G, pos, with_labels=False, node_color='skyblue', node_size=1500, edge_color='gray', linewidths=1)

# Etiquetar los nodos con los nombres de las empresas
labels = nx.get_node_attributes(G, 'label')
nx.draw_networkx_labels(G, pos, labels=labels)

# Etiquetar las aristas con los pesos
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title('Grafo Valorado de Correlación')
plt.show()
```

#### Creamos nuestro operador con los pesos de las arista y lo introducimos como entrada en el algoritmo QA0A

```python
# Problem to Hamiltonian operator
# BBVA - SAN - GOOG - GOOGL
lista_coovarianzas = [("ZZII", 0.86), ("ZIZI", 0.37), ("ZIIZ", 0.36), ("IZZI", 0.37), ("IZIZ", 0.37), ("IIZZ", 0.99)]

hamiltonian = SparsePauliOp.from_list(lista_coovarianzas)
# QAOA ansatz circuit
ansatz = QAOAAnsatz(hamiltonian, reps=2)

ansatz.decompose(reps=4).draw(output="mpl", style="iqp")
```

```python
ansatz.decompose().draw(output="mpl", style="iqp")
```

#### Seteo mi cuenta y simulo un backend real

```python
# To run on hardware, select the backend with the fewest number of jobs in the queue
service = QiskitRuntimeService(channel="ibm_quantum", token="7208e501973a349de8f4ab38efb2ffdb9cb9532d2571a80d3b836241bbc0d95afb4f5477f29cf2c3f31bb99551fb48bb4f9c8491e6aea27f2442f6c28a86135f")
backend = service.least_busy(operational=True, simulator=False)
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime import QiskitRuntimeService
backend = service.get_backend("ibm_kyoto")
aersim_backend = AerSimulator.from_backend(backend)
```

#### Transpilo el circuito a el backend que he elegido para poder adaptarlo

```python
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

target = aersim_backend.target
pm = generate_preset_pass_manager(target=target, optimization_level=3)

ansatz_isa = pm.run(ansatz)
```

```python
ansatz_isa.draw(output="mpl", idle_wires=False, style="iqp")
```

#### Observo mi operador final y su distribución

```python
hamiltonian_isa = hamiltonian.apply_layout(ansatz_isa.layout)
hamiltonian_isa
```

#### Defino mi función de coste

```python
def cost_func(params, ansatz, hamiltonian, estimator):
    """Return estimate of energy from estimator

    Parameters:
        params (ndarray): Array of ansatz parameters
        ansatz (QuantumCircuit): Parameterized ansatz circuit
        hamiltonian (SparsePauliOp): Operator representation of Hamiltonian
        estimator (EstimatorV2): Estimator primitive instance

    Returns:
        float: Energy estimate
    """
    pub = (ansatz, [hamiltonian], [params])
    result = estimator.run(pubs=[pub]).result()
    cost = result[0].data.evs[0]

    return cost
```

#### Creo mi session y cuanto shots quiero hacer

```python
from qiskit_ibm_runtime import Session, SamplerV2 as Sampler

session = Session(backend=aersim_backend)

# Configure estimator
estimator = Estimator(session=session)
estimator.options.default_shots = 100000
estimator.options.dynamical_decoupling.enable = True

# Configure sampler
sampler = Sampler(session=session)
sampler.options.default_shots = 100000
sampler.options.dynamical_decoupling.enable = True
```

#### Lo ejecuto de forma iterativa modificando los parámetros

```python
x0 = 2 * np.pi * np.random.rand(ansatz_isa.num_parameters)
res = minimize(cost_func, x0, args=(ansatz_isa, hamiltonian_isa, estimator), method="COBYLA")
# Assign solution parameters to ansatz
qc = ansatz.assign_parameters(res.x)
# Add measurements to our circuit
qc.measure_all()
qc_isa = pm.run(qc)
qc_isa.draw(output="mpl", idle_wires=False, style="iqp")
```

```python
result = sampler.run([qc_isa]).result()
samp_dist = result[0].data.meas.get_counts()
# Close the session since we are now done with it
session.close()
```

#### Muestro los resultados

```python
plot_distribution(samp_dist, figsize=(15, 5))
```
